## 全球半导体材料行业市场规模整体呈上升态势
- 2019 年全球半导体材料行业市场相比 2018 年下降 1.12%，主要系 2019 年**下游增速放缓，叠加贸易摩擦**，使得半导体产业整体低迷，增速下滑
- 2020 与 2021 年**由于 5G 和新能源的快速发展**，大幅提升了半导体产业的市场需求，半导体材料市场规模快速上升
- 晶圆制造材料占比提升（封装材料占比下降）的原因
  - 先进制造的持续发展，**对晶圆制造环节的材料提出了更高的要求**
  - 加工工艺步骤的不断增加也**提升了晶圆制造材料的消耗量**

  - **封装测试业**

  - 包括

# <font color='red'>半导体硅片</font>
- 半导体器件的主要载体，集成电路大厦之地基
- 硅基半导体材料是目前产量最大、应用最广的半导体材料
- 多晶硅是单质硅的一种形态
  - 按照纯度不同，将**多晶硅**分为
    - **工业级**（纯度在 95%-99% ）
    - **太阳能级**（纯度在99.99%-99.9999%）
    - **电子级**（纯度在 99.9999999%-99.999999999%）
  - **电子级的多晶硅**经过**拉单晶硅锭**、**切割**和后续一系列**加工**后，成为**半导体硅片**

- 根据工艺，半导体硅片可分为
  - 研磨片
    - 可用于制造分立器件
  - 抛光片
    - 轻掺抛光片可用于制造大规模集成电路，或作为外延片的衬底材料
    - 重掺抛光片一般用作外延片的衬底材料
    - 相比研磨片，抛光片具有更优的表面平整度和洁净度
    - 在抛光片的基础上，可以制造出退火片、外延片、SOI 硅片和结隔离硅片
      - 退火片在氢气或氩气环境下对抛光片进行高温热处理，以去除晶圆表面附近的氧气，可以提高表面晶体的完整性
      - 结隔离硅片是根据客户的设计，利用曝光、离子注入和热扩散技术在晶圆表面预形成 IC 嵌入层，然后再在上面生长一层外延层
  - 基于抛光片制造的特殊硅片外延片
    - 是在抛光片表面形成一层**气相生长的单晶硅**
    - 可满足需要晶体完整性或不同电阻率的多层结构的需求
  - SOI硅片（Silicon-On-Insulator）
    - 是在两个抛光片之间插入高电绝缘氧化膜层
    - 可以实现器件的高集成度、低功耗、高速和高可靠性
    - 在活性层表面也可以形成砷或砷的扩散层
- 应用最广的三类硅片是
  - 抛光片
  - 外延片
  - SOI硅片
- 半导体硅片可分为正片、假（陪）片
  - 正片用于半导体产品的制造
  - 假片用来暖机、填充空缺、测试生产设备的工艺状态或某一工艺的质量状况
    - 假片一般由晶棒两侧品质较差部分切割而来，由于用量巨大
    - 在符合条件的情况下部分产品会回收再利用，回收重复利用的硅片称为可再生硅片

## 半导体市场推动硅片市场增长，大尺寸硅片需求提升
- 2016-2018 年，全球半导体硅片市场总体处于上升增速明显
- 2018-2020 年小幅回落
- 2021 年重拾升势，全球市场规模 126 亿美元
- 中国国内半导体硅片市场
  - 自 2015 年起迅速上升
  - 2021 年中国市场规模 16.6 亿美元
  - 2015-2021 年 CAGR 为 27%，市场增速明显
- 历史上半导体行业的年均增速高于电子系统整体市场，主要驱动力是**电子系统中使用的半导体的含量不断增加**
- **集成电路“<font color='red'>制程</font>缩小”推动硅片向大尺寸发展**
  - 先进制程占比上升，推动 12 英寸芯片需求提升
  - 2021-2024年，全球芯片制造产能中 **10nm 以下制程**占比迅速上升
- 硅片的应用与硅片尺寸密切相关
  - 12 英寸硅片通常用于集成电路先进制程，例如智能手机处理器、电脑的 CPU 和显卡、人工智能、储存芯片等高新领域
  - 8 英寸硅片主要制造电源管理器、MEMS、显示驱动与指纹识别的芯片。无论是物联网，还是汽车电子都需要它的技术支撑
  - 6 英寸硅片则主要涉及功率器件、分立器件等常用领域
  - 越是先进工艺的芯片，其所用的硅片尺寸也越大
- 根据尺寸大小的不同，硅片可分为
  - 50mm（2英寸）
  - 75mm（3英寸）
  - 100mm（4英寸）
  - 150mm（6英寸）
  - 200mm（8英寸）
  - 300mm（12英寸）
  - 英寸为硅片的直径
  - 目前**8英寸和12英寸硅片为市场最主流的产品**
- 为何芯片越先进所需硅片越大？
  - 溯其根本，只有两个字——成本
  - 芯片是矩形，这是基于成本的考量
  - 晶圆是圆形，也是为了控制成本
  - 只要晶圆足够大，芯片足够小，拼起来的边缘就会逐渐向圆形逼近（微积分原理）
- 基于成本考虑，分立器件继续沿用小尺寸，集成电路向大尺寸迁移
  - 分立器件由于价格偏低，生产厂商对于投资大尺寸产线动力不足，目前仍以 6 英寸及以下硅片为主。
  - 集成电路使用大尺寸硅片带来的经济效益明显
  - 若硅片尺寸增大带来的成本节约可以弥补投资大尺寸晶圆制造产线的成本，厂商便有向大尺寸迁移的动力

# <font color='red'>掩膜版</font>
- 图形转移传送带，光刻复制的蓝本
- 掩膜版是微电子制造过程中的图形转移母版
- 掩膜版的作用是将设计者的电路图形通过曝光的方式转移到晶圆上，从而实现批量化生产
  - 作为光刻复制图形的基准和蓝本
  - 掩膜版是连接工业设计和工艺制造的关键
  - 掩膜版的精度和质量水平会直接影响最终芯片产品的优品率
- 晶圆制造过程需要经过**多次曝光**工艺，利用掩膜版的**曝光掩蔽作用**，在半导体晶圆**表面**形成**栅极**、**源漏极**、**掺杂窗口**、**电极接触孔**等
- 掩膜版的主要原材料包括
  - 掩膜基板
  - 光学膜
  - 化学试剂
  - 包装盒等辅助材料
- 掩膜版主要应用于**平板显示**、**半导体**、**触控**和**电路板**的制造过程
  - 半导体
    - IC制造
    - IC封装测试
    - 半导体器件
    - LED芯片
  - 平板显示、半导体等中游电子元器件厂商的终端应用主要包括消费电子（电视、手机、笔记本电脑、平板电脑、可穿戴设备）、家用电器、车载电子、网络通信、LED 照明、物联网、医疗电子以及工控等领域
- 光掩膜生产工艺流程主要包括
  - CAM 图档处理
  - 光阻涂布
  - 激光光刻
  - 显影
  - 蚀刻
  - 脱膜
  - 清洗
  - 宏观检查
  - 自动光学检查
  - 精度测量
  - 缺陷处理
  - 贴光学膜等环节

# <font color='red'>光刻胶</font>
- 集成电路制造之纽带，光刻刻蚀衔接链
- 是利用光化学反应进行微细加工图形转移的媒体
- 对光敏感的感光材料，主要成分
  - **成膜剂（成膜树脂）**：一种惰性聚合物基质，是用来将其他材料聚合在一起的粘合胶，占比 10%-40%
  - **光敏剂（光引发剂）**：固态物质，是核心部分，在特定光辐射能下会产生化学反应（改变成膜树脂在显影液中的溶解度），占比 1%-6%
  - **溶剂**：主要起溶解作用，将光引发剂与添加剂溶解形成液态，一边均匀涂抹在器件表面，占比 50%-90%
  - **添加剂**：是固态物质，分为单体和助剂，主要对光化学反应和整体性能起调节作用，占比小于 1%
- 光刻胶按其形成的图像分类有两大类
  - **正性**：在光刻胶工艺过程中，涂层曝光、显影后，**曝光部分被溶解，未曝光部分留下来**，该涂层材料为正性光刻胶
  - **负性**：如果**曝光部分被保留下来，而未曝光被溶解**，该涂层材料为负性光刻胶
- 光刻胶按照用途主要分为三类
  - 半导体用光刻胶
    - 光刻胶根据对应波长，主要品类分为
      - 紫外光谱（300-450nm）
      - g-line（436nm）：g-line、i-line 广泛应用于 0.5um 以上和0.5-0.35um 制程
      - i-line（365nm）
      - KrF（248nm）：KrF 多应用于 0.25-0.13um
      - ArF（193nm）：ArF 应用于 130-7nm
      - EUV（13.5nm）：EUV 应用于7nm 及以下
  - 平板显示用光刻胶
    - 彩色、黑色光刻胶：制备彩色滤光片
    - LCP/TP衬垫料光刻胶
    - TFT-LCD中Array用光刻胶
  - PCB光刻胶
    - 干膜光刻胶：微细图形加工
    - 湿膜光刻胶
    - 光成像阻焊油墨

# <font color='red'>湿电子化学品</font>
- 保证工艺精度的重要材料，国产替代任重道远
- 又称**工艺化学品**，主要用于**半导体、光伏太阳能电池、LED 和平板显示**等电子信息产品的**清洗、蚀刻**等工艺环节
- 是微电子、光电子湿法工艺（主要包括湿法刻蚀、湿法清洗）制程中使用的各种液体化工材料
- 湿电子化学品按照大类一般可划分为
  - 通用化学品（通常为超净高纯试剂）

# <font color='red'>CMP材料</font>
- 抛光液&抛光垫，平坦化技术核心材料
- CMP 是一种化学腐蚀和机械研磨相结合的平坦化半导体表面工艺，是集成电路晶圆制造中实现晶圆全局均匀平坦化的关键工艺
- 半导体制造分为
  - 前道工艺：在晶圆上**形成器件**的工艺过程，也称晶圆制造
    - 前道工艺共有七大工艺步骤，分别为
      - 氧化/扩散
      - 光刻
      - 刻蚀
      - 离子注入
      - 薄膜生长
      - 清洗与抛光
      - 金属化
    - 通过循环重复上述工艺，最终在晶圆表面形成立体的多层结构，实现整个集成电路的制造  
  - 后道工艺：将晶圆上的**器件分离，封装**的工艺过程
- CMP 工艺过程中涉及的耗材包括：
  - 抛光机
  - 抛光液
  - 抛光垫
- CMP 抛光材料在集成电路制造材料成本中占比 7%
  - 其中 **CMP 抛光垫**、**CMP 抛光液**、**CMP 清洗液**分别占比 33%、49%、5%
  - 合计占 CMP 抛光材料成本的 85%以上

## 制程缩小提高 CMP 次数，封装工艺进步拓宽 CMP 边界
- 晶圆制程缩小需大幅提高 CMP 次数，导致 CMP 抛光耗材在晶圆制造过程中消耗量增加
- CMP 的应用边界，从最初的 STI（浅沟槽隔离层）拓展到 ILD（层间介质）、Metal（金属互连层）、TM（顶层金属）等
- 逻辑芯片正向 7nm 以下先进制程发展
  - 芯片制程从成熟制程 28nm，先进制程 14nm 上升到 7nm 后，CMP抛光步骤大幅增加
- 存储芯片的封装工艺进步，让 CMP 工艺从前道延展到后道
  - 存储芯片由 2D NAND向 3D NAND 技术变革，也会使 CMP 抛光步骤数近乎翻倍
- TSV 硅通孔技术作为一项高密度封装技术也需要用到 CMP
  - TSV 正在逐渐取代目前工艺比较成熟的引线键合技术，被认为是第四代封装技术
  - CMP 抛光材料将在先进封装工艺中寻找到新的市场空间

# <font color='red'>靶材</font>
- PVD 沉积核心材料，薄膜沉积重要组成部分
- 物理气相沉积（Physical Vapor Deposition）技术是制备电子薄膜材料的主要技术之一，是利用物理方法**在基板表面沉积薄膜**的方式，根据沉积方式的不同，PVD 分为**溅射法**和**蒸镀法**，**被沉积的材料称为靶材**
- **溅射法**
  - 利用**离子源**产生的离子，在真空中加速聚集成**高速离子流**，轰击固体表面
  - 离子和固体表面的原子发生动能交换，**使固体表面的原子离开靶材并沉积在基材表面**，从而形成纳米（或微米）薄膜
  - 被轰击的固体是 **PVD 沉积薄膜**的原材料，称为**溅射靶材**
- **真空蒸发镀膜**
  - 指在真空条件下，利用**膜材加热装置（称为蒸发源）**的热能，通过加热蒸发某种物质使其沉积在基板材料表面的一种沉积技术
  - 被蒸发的物质是用真空蒸发镀膜法沉积薄膜材料的原材料，称之为**蒸镀材料**
  - 真空蒸发镀膜技术具有简单便利、操作方便、成膜速度快等特点，主要应用于小尺寸基板材料的镀膜
- 按下游应用可分为
  - 半导体靶材
  - 平面显示靶材
  - 太阳能靶材
  - 其他类
- 按形状不同可分为
  - 长靶
  - 方靶
  - 圆靶
- 按照化学成分不同，可分为
  - 金属靶材
  - 化合物靶材
  - 合金靶材
- 靶材产业链可以分为四个环节
  - 金属提纯
    - 原材料铝、铜、钽、钛等金属以金属提纯方式形成高纯金属，作为靶材制造的原材料
  - 靶材制造
    - 将高纯金属通过加工形成溅射靶材
    - 制造好的靶材包括两部分
      - 靶坯：靶坯是溅射靶材的主体
      - 背板：背板起固定靶坯的作用
  - 镀膜（以溅射镀膜为例）
    - 以高速离子束流轰击靶坯，溅射出靶坯表面原子，**沉积于基板**从而**制成电子薄膜**
    - 电子薄膜按照应用不同有不同分类
  - 终端应用
    - 半导体芯片
    - 平板显示器
    - 信息存储
    - 光学元器件
    - 薄膜太阳能
- 靶材市场主要分布于四大领域
  - 平板显示
  - 记录媒体
  - 太阳能电池
  - 半导体：占比约 10%
