# 中国医药行业未来发展关键


## 融入全球监管和卫生系统
- 一个挑战是提供满足国际监管要求的临床数据
  - 要求能否反映其所在地区的患者群体人口统计学要求的临床数据，这就限制了正在中国开展的临床试验的价值
- 一个挑战是关乎数据和严谨性
  - 中国的临床实践和试验数据必须完全符合 FDA 和 EMA 的数据披露和数据安全要求、并且考虑具体适应症标准治疗的变化

## 国内准入环境的变化
- **创新药**仅获批中国上市不足够的，还必须为患者所用。
- 这可以通过三种方式实现。
  - 第一、药物未满足的需求非常高以至于患者愿意自掏腰包支付。
  - 第二、药物可以被商业健康保险计划覆盖。
  - 第三、这是最重要的，药物可以列入国家报销药品目录（NRDL）。
- 中国现在每年更新国家医保目录，这是个好消息。但创新药的回报还不确定是否足以可覆盖本地生物技术公司日趋上升的研发投资

## 上游创新能力
- 中国在**生命科学**和**相关学术人才库**方面取得了良好进展
- 中国在科学突破的速度和规模方面仍面临障碍，并且缺乏成熟的**生态系统机制**，难以将这些研究和知识产权进展**转移到生物制药管线**中

## 组织的适应性
- 如果中国的**生物制药企业**要成功地成为真正具有全球性企业，能够从不同国家的投资者那里获得资金，并在想去的地方经营，就必须调整自己的运营模式，接受多样化的工作实践，并吸引来自世界各地的人才

## 估值的方法论
- 有的是化工属性
- 有的是精密制造属性
- 有的是科技属性
- 有的是消费属性
- 有的是服务属性

### 化学药
  - 是从天然矿物、动植物中提取的有效成分，以及经过合成而制得的药物，统称为化学药物。
  - 分子量一般在1000Da以下，也可以看作小分子药物。
  - 总体来说合成技术工艺成熟，结构明确稳定，具有预防、治疗、或为了调节人体功能、保持身体健康的功效。
  - 化药里又进一步分为创新药和仿制药。
  - 如果简单区分，一般我们**口服的都是化药**

### 生物药
  - 利用生物体、组织、细胞、体液等制造的一系列治疗制品
  - 分子量一般在1000Da以上，即大分子药物
  - 生产技术工艺相对于化学药而言，更为复杂，技术壁垒较高。
  - 生物药里分为生物创新药，和生物类似药
    - 生物类似药基本等同于化药的仿制药，之所以叫类似药，是因为化药的仿制药是可以做到几乎100%一致的，而生物药做到这一点是很难的，所以只能说是类似药。

### 疫苗
  - 主要分为免疫规划类的，和非免疫规划类的。
  - 免疫规划类
    - 就是政府出钱我们打的那些疫苗，比如孩子小时候打的乙肝疫苗、卡介苗、百白破、麻腮风之类的；
  - 非免疫规划类
    - 就是我们自己选择付费打的一些疫苗，比如狂犬疫苗，HPV之类的。
    - 相对来说非免疫规划类的投资机会更大，潜在商业空间更大

### 血制品
  - 白蛋白、免疫球蛋白、凝血因子等。
  - **目前国内企业主要以白蛋白为主**。
  - 和单纯的献血不同
    - 献血是血液输入
    - 血制品是血液经过提炼加工，属于药品

### 医疗器械
  - 分四种：医疗设备、高值耗材、低值耗材、IVD。
  - 医疗设备
    - 医院里会用到的一些设备仪器，进一步细分，可以分为
      - **诊断类设备**
        - 如：CT机，核磁， X射线等
      - **治疗类设备**
        - 如：放疗仪器等
    - 从是否可能会对患者有负面伤害以及设备金额分为
      - 甲类设备
      - 乙类设备
  - 高低值耗材
    - 小到注射针头，大到心脏支架，主要是根据价值区分
    - 但没有特别明显的划分金额，总的来说金额大的就是高值耗材，金额小的就是低值耗材。
  - IVD
    - 体外诊断产品，比如我们进行血液、尿液、粪便、身体组织化验确定是否存在疾病用到的相关消耗品和仪器。
    - 常见的IVD包括：血糖试纸、妊娠检测棒、肿瘤标志物检测试剂盒、hiv检测试剂盒等
    - 和诊断类设备的区别
      - 诊断类设备比如CT我们是不需要从身体采样
      - IVD是需要我们采样化验，比如我们的核酸检测就是IVD相关

    - 不需要医生处方，不在医生指导监督下使用；
    - 如健胃消食片，龙牡壮骨颗粒之类的就是OTC药品
    - 相关企业如：华润三九、华森制药、葵花药业等。

  - 器械OTC
    - 比如鱼跃医疗，他的一部分业务，血压仪、制氧机、血糖仪、雾化器之类的，网上就能买，买了不用专业医生指导就能用；

  - 品牌中药及保健品
    - 东阿阿胶、片仔癀等；

## 总结
  - 上游偏科技
    - 越偏向上游因为和创新相关，业绩弹性更大，但不确定性也较高
    - 越偏向上游变化越多越复杂对专业知识的要求也更高
  - 中游偏制造
  - 下游偏消费
    - 越偏向下游更加稳定，但可能想象的空间就偏少
    - 越偏向下游就可以找到一些好分析的企业，比如眼科、医美、口腔，一些中药保健品之类的
    - 贴近普通人消费，政策管控相对就会更严格
      - 如近些年集采的仿制药、一些特别高溢价的创新药，一些耗材
        - 医保类目的必须品，价格控制会更严格
        - 医美之类的可选消费，可能受到的影响就偏小

# 上游 - 与药相关

## 化工原料
- 医药行业与化工行业密切相关，许多医学产品都是由**化合物**制成的。
- 以下是与医药相关的一些重要上游化工原料：
  - <font color="red">氨基酸</font>：氨基酸用于制造各种**蛋白质**和**抗生素**等药物。
  - <font color="red">抗氧化剂</font>：抗氧化剂在制造**维生素、激素**和其它药物中扮演着重要角色。
  - <font color="red">醇</font>：**乙醇，丙二醇，丙三醇**等醇类物质是许多药物和**疫苗制备的主要原料**。
  - <font color="red">酶</font>：各种酶是医药行业中理想的**催化剂**，可用于制备各种**生物活性化合物**。
  - <font color="red">反式物</font>：如**双酚A**，**多臂聚氧核苷酸**等，用于**制造药物、塑料和电子元件**等。
  - <font color="red">有机酸</font>：**单宁酸**、**水杨酸**等有机酸用于制造**抗生素、防腐剂**和**染发剂**等药品。
  - <font color="red">溶剂</font>：很多药物和化妆品需要用到溶剂，如**甲醇、乙醇、环己酮、丙二醇**等。
  - 除此之外，制造医药产品还需要使用一些其他的化工原料，如各种
    - <font color="red">试剂</font>
    - <font color="red">添加剂</font>
    - <font color="red">催化剂</font>等。

## 原料药

- 由**化学合成**、**植物提取**或者**生物技术**所制备的各种用来作为药用的**粉末**、**结晶**、**浸膏**等用于生产各类制剂的原料药物
- 病人**无法直接服用**

### 分类与子类

- 子类
  - 大宗原料药
  - 特色原料药
  - 辅助用药

### 投资逻辑
- <font color="red">大宗原料药</font>
  - 类似于**成熟化工产品**，**市场整体成熟**，**下游需求增长缓慢**，而且有一定的波动性，具有周期属性
    - 因为投资逻辑偏于周期
      - **集中采购政策**带来的新机会
        - 在整个集采背景之下很多药品价格降幅较大，对原料药的成本控制就显得越发重要
          - 原料药产能的释放需要较大的资金和设施投入，呈现出规模效应
          - 给整个原料药板块带来了一些新的契机
      - **环保限产**淘汰掉了大量的不规范产能，行业壁垒的变高也为头部企业带来了契机
  - 不涉及专利技术。产品如**维生素类**，**抗生素类（青霉素）**等
  - 由于创新药的专利到期，将导致仿制药的品种与数量迅速上升
    - 因此会为原料药行业带来了巨大的市场机遇
  - <font color="blue">价格波动主导股价走势</font>
    - 这个赛道里的波动往往来自于一些突发事件所带来的竞争格局变化
      - 比如生产事故、环保限产等因素引发的供给端收缩，使得供需格局出现变化，从而给头部企业带来较好的收益
      - 对于这类企业，相对于PE来说，PB/ROE是更好的估值指标，体现了周期股的属性
- <font color="red">特色原料药</font>
  - 跟**药品专利**密切相关
    - 因为特色原料药是指**处于专利保护期**的药品的原料药
    - 或**处于专利保护期结束后一段时间内**的药品的原料药
  - 特色原料药**利润率**通常要**高于**大宗原料药
  - 特色原料药的整体特征
    - 因为和专利密切相关，所以专利到期时，**进入市场快**的公司能够尽早享受相关的利润
    - 同样因为特色原料药的高利润特征，导致参与仿制的公司很多，所以导致**产品的价格不断的下降**
  - 特色原料药包括**抗高血压**、**抗肿瘤**、**降血糖类**等，下游对应着专利药或仿制药、特色药物，产品往往具备一定的附加值
  - 行业具备一定的成长性
    - 成长驱动力主要来自于下游药品的生命周期
      - 如：下游药品生命属于上升期，产品放量中，意味着上游原料药也会有相应的利好和放量
      - 因此，在投资实践中，需要跟踪研究其下游产品的情况
  - 这类企业的盈利模式，通常是与一些大企业建立长期稳定的合作关系
    - 下游产品放量，上游会买更多原料药
- <font color="red">辅助用药</font>
  - 辅助用药和**中药注射剂**，现在受到**医保控费**的影响，风光不再，处于萎缩状态

- <font color="blue">原料药包含了周期品的逻辑</font>
  - 目前来说，整个原料药行业产能是过剩的
  - 需要紧盯原料药价格
  - 投资原料药的逻辑在于行业整合，分散的市场竞争经过整合后，形成寡头垄断的格局，让企业有了定价权，价格也是节节高，盈利暴涨
  - 未来原料药行业转移和行业内整合，甚至是原料药企业向下游制剂的转型，也许会成为常态
- 原料药企业的中长期成长逻辑
    - 如果仅仅依赖于下游产品放量，很难驱动企业中长期成长
    - 因此，转型是原料药企业的必经之路。
      - 如以前只做原料药，后续也开始做制剂
      - 或者更进一步转型，做 CDMO 帮别人生产
    - 集中采购政策带来新机会
      - 集采背景之下很多药品价格降幅较大，对原料药的成本控制就显得越发重要
      - 原料药产能释放需要较大的资金和设施投入，呈现出规模效应
    - 环保限产淘汰掉大量的不规范产能，行业壁垒的变高为头部企业带来了契机

  - 技术越简单，产量相对较高，价格越低竞争也越激烈，毛利率也越低

# 中游 - 与药相关

## 化学制剂（小分子药）
- 化学制剂是指从**天然矿物**、**动植物**中提取的有效成分，以及经过**合成**而制得的药物，一般我们口服的都是化药。
- 其中
  - **创新药**是指**在已有药物的基础上**做一些改进并同时**拥有自主知识产权专利**的药物
  - **仿制药**是指与**原研药**具有相同的活性成分、剂型、给药途径和治疗作用的药物（**没有自主知识产权**）

### 分类与子类
  - 仿创药
    - 仿制药
    - 创新药
  - OTC
    - 场外交易（over-the-counter），非处方药
    - 可在药店或超市等零售场所自由出售的药品，无需医生处方即可购买和使用
    - 这些药品通常用于治疗常见的轻微疾病、症状和不适，如感冒、咳嗽、头痛、发热、胃痛、皮肤问题等
  - 普药

### 投资逻辑

- <font color="red">仿创药（仿制药，创新药）</font>
  - <font color="blue">仿制药投资逻辑</font>
    - 增长的确定性比较强，有的增长20-30%，甚至更高
      - 因为前期都是**仿制进口药**，仿制完了之后就**进口替代**，所以增长确定性强。
    - 由于**带量采购**政策影响，期间仿制药板块估值位于底部
      - 已不太需要销售，医院绑定了销售量
      - 整个产品降价幅度较大，从过去高毛利逐步变成了低毛利，因此现有产品线和整个销售能力变得不太重要
    - 对于仿制药企业，需要重点分析在研管线，看未来的发展方向
      - 一是转型创新
        - 但是创新药的研发周期是5-8年，需要布局研发团队，研发投入较高
      - 二是发展难仿药
        - 如呼吸剂，产品专利已经过期了，但由于本身剂型上的难点，具备一定**生产壁垒**
      - 三是延伸产品线
        - 上市一款仿制药短期会面临集采，可以带来增量
        - 只要每年保证产品上市的梯队，未来也可以实现较为稳定的现金流和盈利
  - <font color="blue">创新药偏向科技股的逻辑，同时销售能力也是重要的考虑因素</font>
    - 创新药碾压式的疗效往往会导致极强的定价能力，而极强的定价能力往往导致阶段性的很高的ROIC（资本回报率Return on Invested Capital，简称ROIC）
    - 对于创新药企业来说，当产品尚未上市时，PE估值方式并不适用
      - 创新药在上市之前，市场开始有预期的时候，股价就会快速提升
      - 此时产品还没有上市，PE开始上升，市值与PE同步上升
      - 当创新药开始放量的时候，PE和市值可能出现分化
      - 当药品逐步进入成熟期和稳定期，股价开始修复，PE也会回到比较稳定的状态
    - 创新药毛利率比较高，通常是70%以上，净利率可以达到30%以上
    - 一般通过**医药渠道**销售，有自己的销售队伍。（如恒瑞、恩华等）
  - **研发投入很大**，是最重要的制药企业类型
    - 评估一个创新药企业，相比中药和化学仿制药而言，核心在于对**研发管线**的估值
    - 其他考虑因素
      - 临床数据
      - 推进时间节点
  - 估值 - DCF估值法折现
    - 市场是把较多研发管线的估值纳入了当前的市值中
    - 这是为什么恒瑞当前估值绝对值较高的原因
    - 其他考虑因素
      - 企业的销售能力强不强
      - 现有管线能不能稳定成长
      - 有没有现金流
  - **CRO 创新药**
    - 创新药投入高，研发周期长，且很大程度上依赖于海外市场
    - 2019-2021年创新药爆发原因
      - 美元宽松背景下的海外需求的扩张
      - 中国本土创新药从零到一的巨大红利
    - 2022年至今
      - 全球经济下行背景下海外市场需求的萎缩
      - 国内创新药行业也迎来一个泡沫出清时期
      - 美国对中国创新药的政策打击
      - 令CRO赛道急速下坠，业绩增幅下滑，股价下跌
  - 创新药天然是一个**逆集中**的行业，颠覆性的迭代多
    - 创新药之间的竞争本质上是IP间的竞争，对单个或几个人的依赖特别强
    - 现在已经超过3/4的新药最初来自小药企
  
  
  
- <font color="red">OTC</font>
  - <font color="blue">OTC包含消费品（快速消费品）的逻辑</font>
    - 分析消费品需要关注的核心点
      - 量价关系
      - 管理团队
      - 原材料因素
  - 销售**渠道**主要通过**医院**、**零售药店**销售
  - **广告**是营销的主要手段
  - **品牌**体现公司的价值
    - 譬如三精，江中、华润三九等
- <font color="red">普药</font>
  - **毛利率较低**，通常在 50% 以下
  - 没有自己的销售队伍，通过**代理渠道**
  - 增长依托于**产品规模**和产品的**升级换代**，波动性比较大

## 生物制品（大分子药）
- 生物药是指利用**生物体**、**组织**、**细胞**、**体液**等制造的一系列治疗制品。
- 生产技术工艺相对于化学制剂来说，更为复杂，技术壁垒较高。
- 生物药领域是未来的主要方向


### 分类与子类
  - 疫苗
  - 血制品
  - 其他生物制品


#### 细胞和基因疗法（Cell and Gene Therapy, CGT）

- CGT技术分类
  - 细胞治疗相关技术
    - 非基因编辑：新型体细胞和干细胞培养技术
      - 角膜和视网膜修复造血干细胞相关治疗
    - 基因编辑：基因改造细胞技术
      - CAR-T 细胞
      - CAR-NK 细胞
      - CAR-巨噬 细胞
      - DC 细胞
  - 基因治疗相关技术
    - 病毒载体编辑：基因替代（补充功能性基因）
      - 非整合性载体，治疗基因独立于靶细胞基因表达
        - 腺细胞
        - 腺相关病毒
        - 疱疹病毒
      - 整合性载体，常用于体外治疗，存在潜在的插入突变风险
        - 慢病毒
        - 逆转录病毒
    - 非病毒载体编辑：基因编辑（敲除或纠正错误基因）
      - 基因敲除
      - 基因纠正

- 体外治疗
  - 借助整合型病毒载体（如慢病毒载体）在体外条件下将基因导入前体细胞或干细胞基因组，伴随细胞分裂将基因传递至子代细胞，再将细胞回输至体内。
  - 体外治疗的代表为细胞疗法

- 体内治疗
  - 借助质粒、病毒载体将功能基因或健康基因转入至宿主细胞内，实现相关基因的持续表达。
  - 常用载体包括腺相关病毒、腺病毒，非病毒载体包括脂质体等。
  - 体内治疗的代表为基因疗法

- CGT 治疗领域以**肿瘤**和**罕见病**为主，逐渐向其他疾病领域拓展

- **CGT 三大潜在突破方向**
  - **定制化→通用化**
    - 通用型 CAR-T 可采用健康供体的 T 细胞，可进行大规模生产。
    - 可使成本降低为现在的约 1/10，是目前细胞疗法最有希望突破的方向。
  - **血液瘤→实体瘤**
    - 若细胞疗法能拓展到实体瘤领域，潜在的受益人群将大幅增加。
    - 根据 WHO 的统计，每年新发血液瘤患者仅占癌症总新发患者的 6%-7%，绝大部分患者罹患实体瘤。
  - **罕见病→常见病**
    - 目前基因疗法以治疗罕见病为主，每个适应症能治疗的患者十分有限



##### 基因治疗

- 基因治疗是一种旨在通过**更改患者基因**来治疗疾病的治疗方式
  - 这种方法利用了现代生物技术手段来直接影响人类基因组
  - 它的主要思想是使用**DNA**、**RNA**和**蛋白质**等分子来修复或替换异常的基因表达
  - 通过**基因添加**，**基因修正**，**基因沉默**等方式修饰个体基因的表达或修复异常基因，达到治愈疾病目的的疗法

- 一种分类
  - 以病毒为载体的基因替代
  - 非病毒载体的基因编辑

- 基因治疗有许多类型
  - 1. 基因增强治疗
  - 2. 基因靶向治疗
  - 3. dna修复治疗
  - 4. 免疫基因治疗
  - 5. 肿瘤抑制基因治疗
  - 6. 基因靶向药物治疗

- 目前基因治疗的临床应用可分为三个主要领域
  - 癌症（Cancer）
    - **血液系统**相关癌症的基因治疗临床试验数量是最多的
    - 其次是**胃肠道**和**神经系统**相关癌症
    - **白血病**、**淋巴瘤**、**黑色素瘤**和**多发性骨髓瘤**是基因治疗领域最常见的癌症类型
  - 遗传性疾病（Genetic Disease）（单基因或多基因疾病）
    - 遗传性疾病的基因治疗集中的疾病
      - 代谢障碍类疾病
      - 眼科疾病
      - 凝血障碍类疾病
    - 临床试验数量最多的遗传性疾病主要有SMA、DMD、B型血友病、ADA-SCID和β-地中海贫血
    - AAV是最常用的基因治疗载体
  - 传染病（Infection）
    - AIDS、COVID-19、疟疾、埃博拉、丙型肝炎、HPV和乙肝是临床试验数量最多的传染病
    - 从获批的基因治疗产品数量上来看，基因治疗在传染病的应用上还有很长的路要走

- 基因治疗的载体
  - 理想的基因治疗载体应具有的特性
    - 高转导效率
    - 诱导长期稳定表达的能力
    - 靶细胞特异性
    - 负载容量特性
  - 最常见的载体类型是**病毒载体**（病毒载体具有更高的转导效率、更强的工程化能力和高度特异性的基因递送特点）
    - 慢病毒和逆转录病毒
    - 腺病毒载体
    - AAV载体
  - 在**非病毒载体**中，最常见的是**RNAs**（siRNA、miRNA、shRNA）和**质粒**


#### 细胞治疗

  - 传统中药
    - 传统中药，中药OTC和中药饮片：类似于较高壁垒的**快速消费品**
  - 中药OTC

### 投资逻辑

- 现状
  - 中药企业大部分**在研产品**相对偏少，与市场关注的创新方向有一些脱离

- 具备较好的资源禀赋
  - 中药消费品，如东阿阿胶、片仔癀、云南白药等
    - 片仔癀
      - 具有独家配方
      - 是OTC产品，本身具备提价能力
      - 关注这类企业的品牌力和溢价能力、管理层的能力、营销渠道、推广能力和横向扩张能力

  - 传统OTC企业相对偏向成熟，过去五年的复合增长率是7%左右，而整个医药行业的增速大约是10-15%，比行业增速慢一些

    - 传统OTC的投资亮点（中药消费品的投资逻辑相似）
      - 主要来自于**提价**带来的重要驱动力，
      - 如果是一些中药消费品，提价能力就更加重要，
        - 意味着未来的成长空间
        - 只要它提一次价，随着价格的增长，公司就可以实现较好的成长
      - 或是有提价可能，或是可以带来增量，若是量价齐升就是更好的投资逻辑
      - 对于品牌突出的中药消费品，在低谷期都是比较好的买入机会
    - 其它OTC行业，需要研判其自身的周期波段机会

  - 现有的产品有哪些，赛道是否足够宽广，产品价格是否有压力
    - 现有产品是否有定价权
    - 现有文号是否存在二次开发或推广营销的机会，以实现较好的增长（比如同仁堂主要聚焦于现有的大品种）
  - 未来的竞争格局怎样，是否会进入医保

    - 是否有独特的生产药方
    - 是否有进入壁垒

  - 关注政策变化
    - 对**中药注射剂**的限制
    - 落后产能被淘汰之后，重要板块可能会重新崛起
  - 中药材的价格波动
    - 中药材的价格波动性比化学药的波动更大，并且占成本的比例也更高一些
    - 中药材价格也是需要关注的点，会影响到中药的成本

## 医疗器械

### 分类与子类

- 设备与耗材（中游）
  - 医院里用到的一些设备仪器，按用途可划分
    - 医疗设备（有的领域如影像设备，短期内仍然难以撼动外资的地位）
      - 诊断类设备
        - CT机
        - 核磁等
      - 治疗类设备
        - 放疗仪器等
    - 医疗耗材（用于诊断、治疗、保健、康复等的消耗性器件材料）
      - 高值耗材
        - 人工骨、心脏支架、义齿等
      - 低值耗材
        - 注射器、口罩、输液器、血袋等
- 制药装备（中游）
- 体外诊断 IVD（中游）
  - 在人体之外，通过对人体的血液等组织及分泌物进行检测，获取临床诊断信息的产品和服务
  - 直接或者间接用于人体的仪器、设备、器具、体外诊断试剂及校准物、材料以及其他类似或者相关的物品
  - 诸如**核酸检测**、**验血**、**验尿**、**身体组织化验**等用到的**相关仪器和耗材**都属于体外诊断
  - IVD（体外检测）中的生化诊断基本实现了国产替代

### 投资逻辑

- 相对特殊一点
  - 尽管披上了医药的皮，但是本质还是偏向于制造业
- 医疗器械的上游和前面医药行业的上游不太一样
  - 医疗器械的上游相对复杂，除了极少数高精尖的精密仪器，**上游行业基本上是充分竞争的**。
    - 其中包括电子仪表、检测设备、集成芯片、外壳材料、生化材料、电子检测元器件、机械设备等等
  - 行业**下游**一般是面对**医院**、**药店**、**制药公司**和**患者本身**

- 医疗器械行业与药品行业产业链的投资逻辑差异较大
  - 主要因为
    - 医疗器械注重改善生活质量、提高诊断检测效率
    - 药品注重治疗
  - 另外就是
    - **医疗器械产品相对小众**，**单个产品体量和市场空间较小**，**产品生命周期较短**
    - 药品一旦研发成功就是重磅产品，而且药品的生命周期相对较长，单个产品的市场空间比较广阔

- 行业未来**发展趋势**应该是对于一些**高端的医疗器械形成国产替代**。

- 体外诊断行业
  - 上游
    - 核心原料供应商：生物原料（如抗原、抗体、酶等）和化学制品等
      - 核心原料质量的优劣决定了体外诊断试剂和检测的质量
      - 抗原、抗体、酶等核心原料的发展在整个产业链中具有关键意义，是产业链核心环节之一
    - 核心元部件供应商：电子元件、光学元件、模具等
  - 中游
    - 诊断试剂、仪器的研发、生产企业以及流通环节的代理商和经销商

    - 目的
      - 优化药品购销秩序，压缩流通环节
    - 商业逻辑

    - 但是两票制暂时还不会取消
    - 由于压缩了流通环节，利好大型医药流通企业，比如国药股份

# 医药行业投研逻辑

## 特点
- 医药行业整体**需求偏向于刚性**大，**弹性小**
- 和宏观经济的相关程度较小
- 具有**防御性强**的特征
  - 由于**防御属性**和**高确定性**的特征
  - 因此当**市场风险偏好弱**，**不确定性强**的时候，**资金喜欢抱团**取暖，也就是俗称喝酒吃药行情
- 医药行业属于**弱周期**

- 医药供给侧有两个趋势
  - 延长创新药的专利保护时间来刺激创新
  - 简化原研专利到期之后仿制药的上市流程和标准，有效控制药价

- 供给侧改革
  - 加大价格较高的创新药纳入医保的比例（我国是10%，美国是70%）
  - 压缩仿制药的利润水分来进行医保控费


## 投资策略
- 最大市值策略：优先分析和投资行业内市值最大的公司
- 马太效应反映社会中普遍存在的一个现象，赢家通吃。
- 当今社会由于信息因素（传播便捷），物流因素（高效物流减少了地理位置的限制），规模效应，更容易形成马太效应，行业集中度会越来越高，龙头会拿走最大的份额和绝大部分利润，第一市值的公司竞争优势明显。
- 投资中如果不清楚买哪只股票，买行业中市值第一的公司基本没错（当然还需要结合公司财务和公司具体的业务综合分析）。

## 选股逻辑
- 特殊的研发模式
  - **核心技术**和**专利保护**成为壁垒阻碍新进入者
  - 获得临床医生的使用以及医保支付的帮助，使**药物有类消费品的稳定品牌属性**
  - 医药企业针对单一科室的销售与研发迭代、对临床医生的心智占领将成为**核心竞争力**
- 稀缺的**要素资产**
  - 以累计研发费用的方式凝结的产品批文
  - 未来形成类似于海外的**平台型公司**后，对终端的销售覆盖和学术推广能力
  - 以销售费用的形式凝结的，难以简单通过利润和收益进行分析
- 资产负债表分析
  - **不再稀缺的资本**以及**难以控制的产能**并不是形成长期壁垒的决定性要素
  - 只有**不可复制的资产**才是长期牛股的核心竞争力
    - 牌照限制（血制品）
    - 天然不可再生/复制的原材料（道地药材基地）
    - 技术脱钩风险下的**稀缺上游技术**
      - 上游制药设备
      - 基因检测领域的测序仪
      - 生物大分子所必需的纯化微球技术
- 医疗领域中高端人才的培养周期长，供给难以快速提高
  - 特定领域的研发人员同样是稀缺的
  - CRO企业的人效有望持续提高，在全球市场占据重要一席。

### 估值
- 估值需要根据不同的公司特点选取不同的估值方法
- **医药企业大多数为成长性企业**，可以采用PE结合PEG的方法，高成长的医药企业往往享有高估值。
- 总体来说，成长性决定公司的估值水平，成长性高的公司享有高估值。

## 研究重点
- <font color="red">医药行业的供给和需求，其中**需求是关键**</font>
  - **医保**的普及能够一定程度上**缓解和释放相关的需求**
  - 落实到公司产品的研究上
    - 应该重点研究**现有产品**和**在研产品**能否产生新的**需求**
    - 公司产品药品的创新和营销都是为了需求
      - **创新**是为了**创造新的需求**
      - **营销**是为了**巩固和抓住现有的需求**

- 推动医药**长期需求**增长的主要因素
  - 人口持续增长、
  - 老龄化比例提高、
  - 一些疾病发病率提高
- 突发疾病属于引起医药的**短期需求**因素
  - 非典
  - 流感等

- 具体研究医药需求，从以下几个方面分析：
  - 1，**大病种**，大品种
  - 2，消费频率（是否是**慢性病**）
  - 3，该领域疾病的**发病率**是否提高
  - 4，**药品市场的增长和竞争情况**
  - 5，各项**政策**，特别是**医保政策**是否会对该类药品的消费有拉动。
  - 6，需求的增长能否落实到**具体公司**

## 成长的驱动因素
- 推动医药行业成长的都和需求有关
  - 1，<font color="red">政策推动</font>
    - 医药行业是受政策影响非常大的行业
  - 2，<font color="red">增量市场</font>
    - 基层医疗机构对医疗器械的需求
    - 县级医院对医疗器械设备更新换代的需求
    - 社会资本进入医疗领域带来的医疗机构对设备的需求
  - 3，<font color="red">需求升级</font>
    - 产品升级推动需求升级
      - 1，技术升级
        - 数字化技术的发展为高精尖医疗设备的研发驱动，新技术新产品层出不穷。
      - 2，消费能力升级
        - 支付能力提升，健康人群对于医疗器械的消费将大大扩充医疗行业的边界
  - 4，<font color="red">进口替代</font>
    - 新药、医疗设备等等**中高端市场国产化进程**逐步深入，缩小与进口产品的差距，通过高性价比和优质服务实现进口替代
  - 5，<font color="red">需求创造</font>
    - 新药的研发（对癌症，疑难杂症的需求，大量资金投入）
    - 新的疗法，互联网技术革命创造的新需求（可穿戴设备，智慧医疗等）

# 医药知识点与概念

## 创新药类型

### FIC（First in Class）
- **首创药物**，指通过寻找全新的**药物靶标**、**作用机制**和**分子结构**，从无到有**逐步合成候选化合物**，通过反复试验筛选，最终发现既满足治疗效果又满足人体安全性（**耐受程度**、**药代动力**）要求的药物。
- 其研究过程可谓大海捞针，投入大，失败率高，但是一旦成功会是药物治疗领域的重大突破，收益也十分巨大，研发投入几十亿美元的重磅创新药一般都是此类型。

### 跟踪性创新药
- Me-too、Me-better、Best in class 类药物，指的是**在FIC的分子结构**上进行修改，得到分子结构和FIC不同，但疗效相近的药物。
- 由于此类药物是在首创新药的基础上进行研究开发，因此也有将这种创新药研发模式称为**快速跟踪性创新药模式**。
- 这类药物根据其实际疗效及安全性，可分为Best in class、Me-better、Me-too等多种类型。
- 由于是在已经研究出来的FIC药物基础上进行修改，**研发风险大大降低**，但是依然需要通过大量反复地临床试验，**研发费用依旧较高**，在国内一般会是几亿至十亿元的量级。

#### BIC（Best in class）
- 即同类最优
- 该类药物比FIC疗效更好或更安全，可解决了**同靶点**无法解决的**临床需求**
- tips
  - 靶点是指在生物医学领域中疾病发生的**关键蛋白质**或**分子**，它们与某些药物作用关联紧密，可以**被这些药物选择性的调控或抑制**，从而达到治疗疾病的目的。
  - 因此，靶点通常是生物医学研究和药物开发过程中的重要对象，通过对靶点的深入研究和了解，有助于寻找药物的作用机制以及提高药物的效果和副作用的控制。
  
#### Me-better
- Me-better类创新药，即临床疗效或安全性优于首创新药物的新药
- Me-better类创新活动，将成为我国创新药研发的主要途径

#### Me-too
- Me-too药物特指**具有自己知识产权**的药物，其药效和同类的突破性的药物相当。
- 这种旨在避开“专利”药物的产权保护的新药研究，大都以现有的药物为先导物进行研究，其研究要点是找到不受专利保护的相似的化学结构。
- 这种研究有时可能得到比原“突破性”药物活性更好或有药代动力学特色的药物。
- Me-too药物开发模式最大的优点是研发成功率高，相对首创药物而言资金和研发实力门槛更低，因此**Me-too药物往往会成为药企切入创新药研发的重要途径**。

#### Me-copy
- Me-copy是目前中国特色的创新药开发模式
- 与国外Me-too药物不同的是
  - 中国的Me-copy创新药并非通过高通量筛选，从HITs开始优化
  - 而是在同靶点已知活性结构基础上进行修饰，做出足以绕开专利、带活性结构修饰的类似物

## 我国的创新药研发步骤
- 首先从“仿制”转型为“创新”，即从研制“**Me-too**”做起，如恒瑞医药的艾瑞昔布是“Me-too”类药物；
- 第二步是“**Me-better**”，围绕原NCE结构基础上进行二次创新，虽然有结构上仿的特点，但更有创造其自身独特优势的地方，素有“国产易瑞沙”之称的浙江贝达药业的埃克替尼、抗肿瘤作用优于同类药物PTK787（诺华和德国先灵联合研发）的恒瑞医药的阿帕替尼都属于此类药物；
- 第三步是“**Best-in-class**”（**BIC**），是“Me-better”里更强的一个，例如辉瑞的立普妥；
- 第四步就是“**First-in-class**”（**FIC**）药物，即全新化合物，同时靶点也是新的，此类药物研发风险极高，但成功者也是有不少，获得诺贝尔奖的青蒿素就属于此类。


## 我国创新药研发的主要途径
- 目前，我国创新药的层次主要处于以仿制为主到仿创结合的阶段，与海外存在一定的差异。
- 海外市场创新药大多为First-in-class类，高成本高风险下享受高收益
  - 通常在专利到期之前销售一路上行
  - 专利到期后仿制药或生物类似物将以低价抢占市场，销售逐渐下滑
- 而国内创新药以Me-too类药为主，缺乏First-in-class类药物

# 合成生物学
- 合成生物学是继“**DNA双螺旋结构的发现**”和“**人类基因组计划**”之后，以**工程化的手段设计合成基因组**为标志的第三次生物技术革命
- 其**实质**是基于**工程学**的原理，通过在试错过程中引入标准化实验手段，按照特定目标设计、修饰及构建合成生物体系，从而推动生物学从标准化、定量化和通用性等角度系统地形成**工程化发展**。这种方式突破了原有的生物自然进化的局限，能够**定向合成出自然界全新的化合物**。
- **酵母**是一种单细胞真菌，它们广泛存在于自然环境中，如空气、土壤和水中。
  - 酵母通常会在繁殖过程中发酵，并**将糖转化为二氧化碳和酒精乙醇等物质**。
  - 这使得酵母在**酿制啤酒、葡萄酒、面包和其他食品**中得到了广泛的应用。
  - 此外，酵母也是科学家进行基因工程和生物技术研究的实验模型。
  - **酵母菌**在自然界分布广泛，主要生长在**偏酸性的潮湿的含糖环境**

## 行业发展驱动因素
- 基因测序时间和成本的持续减少以多种方式推动合成生物学的快速发展
  - **基因组数据库**是**下游合成生物学应用**的重要起点（**蛋白质表达**、**定向进化**和**代谢工程**）
  - 基因测序产生了大量的电子基因序列信息，这对于设计合成基因和生物成分十分重要
  - 低成本的全基因组测序（仅需1000美元）能够更有效地控制**长基因构建体**的质量，这是**基因合成的关键步骤**
- 新的**基因编辑技术**，正在对合成生物学行业产生重大影响
  - 包括
    - 成簇的定期间隔短回文重复序列 (CRISPR) 
    - 类转录激活因子样效应核酸酶 (TALEN) 系统
      - 允许以快速、有效的方式对基因组进行更改，从而改进并优化合成生物学工作流程
- **合成生物学CRO**可以通过设计、制造和测试新型**微生物底盘**来判断是否适合制造特定的产品
  - CRO提供一系列与合成生物学相关的服务，帮助用户开发新的生产工艺
- **上游关键原材料**（如**寡核苷酸**）成本的降低推动了对合成生物学产品的需求
  - 以低成本的可用性增加了应用和客户的数量，从而推动了销售额的增长
- 以精确且具有成本效益的方式编辑基因的能力对于一系列合成生物学市场非常重要
  - 通过设计并改造微生物的代谢通路来更高效并低成本的实现传统化学合成无法完成的生物化学反应，从而实现产品商业化的增值

> **长链 DNA 合成**  
> 是指在**体外**实现长片段 DNA 分子的合成。  
> 这一技术可以用于产生具有特定功能或特殊序列的基因组 DNA 片段，同时也可以用于制备 DNA 探针、引导 RNA 等多种研究和应用领域。  
> DNA 合成一般使用**固相合成**方法，即通过连接单个**核苷酸分子**构建目标 DNA 序列，并逐渐实现 DNA 长链的扩增。  
> 具体而言，该方法利用了**化学**的**脱保护**、**联接**以及**固相纯化**等原理，在不断重复以上步骤下逐渐合成出具有特定长度的 DNA 片段。  
> 长链 DNA 合成的速度和成功率都取决于所使用的合成试剂质量、实验操作者的技能水平，以及化学反应条件的保持稳定等因素。  
> 该领域目前在不断推出新的**化学方法**以提高 DNA 合成效率和质量，以满足更加严格的研究或产业需求。  
> 对于长链 DNA 合成，需按照以下步骤进行：  
> 1. 设计目标 DNA 序列
> 2. 选择相关合成试剂：包括脱保护试剂、活性调节剂、配基等
> 3. 将其中一端固定于合成支持物上，另一端暴露
> 4. 依次加入各种保护基和核苷酸，生成 DNA 链的扩展
> 5. 联接核苷酸至已存在的 DNA 链的末端
> 6. 重复以上步骤，直到达到所需长度
> 7. 使用溶剂和其他方法移除保护组

> - 长链DNA合成应用优势：
>   - 1）可从无到有（0=>1）设计目标基因，
>   - 2）可根据需要定制各类基因突变/改造，
>   - 3）不受限于样本来源、物种间差异、物种内差异（个体多态性），
>   - 4）也不太受限于基因的长度等。

> - 长链DNA合成应用价值：
>   - 1）将有望完全取代目前以传统分子克隆技术为主的实验室工作流程，
>   - 2）节省大量依赖模板构建人工设计基因的时间和科研经费，
>   - 3）彻底改变数以百万计研究人员的科研思路和实验室的科研模式，
>   - 4）极大提升以基因为研究基础的相关学科发展水平，有力推动相关生物产业发展。

>   - 第三代——酶法合成类公司：酶法合成龙头DNA Script
