# 铝概况

- **铝**元素是仅次于**硅**的地壳中第二丰富的金属元素，含量高达8.3%
- 主要以**铝土矿**、**铝硅酸盐矿石**和**冰晶石**的形式存在
- 这决定了上游和中游供应端的充裕

- 铝的重量约为钢或铜的三分之一
- 具有**延展性**、易于加工和铸造
- 具有优异的**耐腐蚀性**和**耐久性**
- **铝**的使用量都超过了除**铁**以外的任何其他金属，而且它在世界经济的几乎所有领域都很重要

- **铝产业链**为：
  - 铝土矿-氧化铝-电解铝-铝深加工-铝终端需求
  
- **氧化铝**按**物理性质**不同，通常将**工业氧化铝**分为**砂型**、**中间型**和**粉型**三种  
  
  

- 铝产业链特征
  - 生产原料单一
    - **铝土矿**是生产**氧化铝**的唯一原料
    - **氧化铝**是生产**电解铝**的唯一原材料
  - 生产工艺单一
    - 全球 90%的铝业公司采用**拜耳法**生产氧化铝
    - **电解铝**全部用**熔盐电解法**生产

# 产业链

## 上游：矿石与中间品

### 铝土矿

#### 铝土矿
- 又称**高铝矾土**
- 铝土矿是一种以**氧化铝水合物**为主要成分的复杂**铝硅酸盐矿石**
- 铝土矿的主要成分有：
  - Al₂O₃
  - SiO2
  - Fe2O3
  - TiO2
  - 少量的CaO、MgO硫化物
  - 微量的镓、钒、磷、铬等元素的化合物
- 工业上能利用的，以**三水铝石**、**一水软铝石**或**一水硬铝石**为主要矿物所组成的矿石统称
  - 铝土矿按照冶炼的**难度由低到高**：三水铝石型、一水软铝石型和一水硬铝石型
- 按照主要组成成分分为三类
  - 三水型铝石
  - 一水型软铝石
  - 一水型硬铝石
- 铝土矿按照其化学成分的氧化铝含量和矿物类型分为六个级别
  - 备注：**铝硅比**是指铝土矿矿石中**三氧化二铝(Al₂O₃)**与**二氧化硅(SiO2)**的百分含量之比，它是衡量铝土矿品质的最主要标准之一，铝硅比愈高的矿石品质愈好
  - 三水铝石化学活性最大、一水软铝石次之、一水硬铝石较弱、**刚玉则是非常稳定的氧化铝**，在不同的冶炼条件下，产生不同结构的氧化铝
  
|级别|氧化铝占比不低于（%）|铝硅比不低于（%）|用途|
|:-:|:-:|:-:|:-|
|一级品|70|12|氧化铝，刚玉型研磨材料|
|二级品|65|7.0-9.0|氧化铝，刚玉型研磨材料，高铝水泥|
|三级品|60|6|氧化铝，高铝水泥|
|四级品|50|3.0-3.5|氧化铝|
|五级品|45|7|氧化铝|
|六级品|40|3.5|氧化铝|


#### 全球铝土矿“储量”分布
- **几内亚**储量占比第**一**，几内亚铝土矿基础资源储量**74 亿吨**，占全球铝土矿基础储量的**23.62%**；
- **越南**储量第**二**，铝土矿基础资源**58亿吨**。占比 **18.51%**；
- **澳大利亚**储量占比第**三**，达 **16.28%**；
- **中国**储量第**七**，查明储量为**7.1亿吨**，占比为 **2.27%**
  - 中国的铝土矿以第三种（一水硬铝石型）为主，需要在高温高压条件下才能溶出，最难冶炼，占全国总资源量的 99%以上，三水铝石型的铝土矿还不到 1%，且品位低、规模小
  - 我国铝土矿资源分布较为集中，主要分布在**广西、河南、贵州、山西**四个省区，储量合计占全国总储量的**91.9%**
    - 广西30.00%
    - 河南24.43%
    - 贵州24.10%
    - 山西13.37%
  - 国内铝土矿**铝硅比偏低**，大多数为**高铝**（Al2O3大于60%）、**高硅**（SiO2在5%~15%之间）、**低铁**（Fe2O3小于10%），且矿体薄，品位波动大。适合经济开采和利用的铝土矿资源少。
  
- 全球铝土矿矿床类型分为两大类
  - 红土型
    - 红土型矿床主要以高铁、低硅、高铝硅比的三水铝石型矿石为主，品质优良，主要分布在赤道附近
    - 三水铝石、一水软铝石
    - 中铝、低硅、高铝硅比、高铁、适用拜耳法
    - 储量88%
    - 全球大多数的矿床为红土型铝土矿，属于三水铝石型和一水软铝石型，占比**88%**，即**高品位矿**
  - 沉积型
    - 沉积型（岩溶型）矿床主要以高铝、高硅、中低铝硅比、低铁的一水硬铝石为主，品质次之，主要分布在温带地区
    - 一水型硬铝石
    - 高铝、高硅、中低铝硅比、低铁
    - 需用烧结-拜耳法
    - 储量 11%

- 随着环保和双碳主题，产量上方空间有限

### 烧碱

- 烧碱可以使铝土矿中的氧化铝转变成**铝酸钠溶液**，后分解析出**氢氧化铝**，经分离、洗涤后煅烧，获得氧化铝
- <font color="red">**氧化铝是烧碱最大的下游品种**</font>
  - 生产1吨氧化铝需要烧碱130kg
  - 烧碱占氧化铝生产成本的16%-20%
  - **烧碱与氧化铝价格高度正相关**

- 中国烧碱产能约占全球总产能的46%，产量约占全球总产量的47.9%，是全球最大的烧碱生产与消费大国。
  - 我国烧碱基本上实现自产自销
  - 烧碱消费区域主要集中于华北、华东地区

### 氧化铝

#### 氧化铝
- 氧化铝，又称**三氧化二铝**，是一种难溶于水的白色粉状或砂状物体，理化性能稳定
- 氧化铝按照不同的**生产用途**可分为两类
  - 冶金级氧化铝
    - 全球超过 **90%** 的氧化铝都是冶金级氧化铝（**95%**）
    - **95%**的氧化铝都是用铝土矿冶炼得到
    - 冶金级氧化铝为熔盐电解法生产金属铝所用的氧化铝，是**生产电解铝的主要原料**
  - 非冶金级氧化铝
    - 占比约**5%**
    - 用于制作各种**耐火砖**、**陶瓷**、**阻燃剂**，广泛用于**国防军工**等各个领域
      - 包括用于红外军事装置、民用航天、军工、高强度激光、条形扫描仪、耐磨轴承的纤维等电学、光学、化学、生物、吸声、热学等领域以及精细陶瓷等
- 铝土矿是生产氧化铝第一大原料，生产**一吨氧化铝**需要约**2.3吨铝土矿**，占氧化铝成本约**37%**

#### 冶炼工艺

- 氧化铝**生产方法**有四种：酸法、碱法、酸碱联合法、热法
  - 工业上几乎全部采用**碱法**生产氧化铝

- 氧化铝的**冶炼工艺**大致可以分为
  - 烧结法
  - 拜耳法
    - 主流冶炼方法（鉴于全球高品位富矿占比较大，冶炼工艺多为拜耳法）
    - 母液可以循环使用

### 电解铝
- 铝的电解是以氧化铝为原料、冰晶石为溶剂的电解过程
- 电解铝的生产原理是
  - 使直流电通过以氧化铝为原料、冰晶石为溶剂组成的电解质
  - 在950~970摄氏度下使电解质溶液中的氧化铝分解出铝
- 从质量守恒定律讲，生产吨铝需要1.89吨氧化铝
  - 考虑到**耗损**，通常认为，**吨铝需耗氧化铝1.92-1.93吨之间**

#### 冶炼工艺：熔盐电解法

- 铝电解的生产过程分3个阶段
  - 焙烧
    - 电解槽的焙烧主要目的是烘干炉体
    - 使电解槽达到**930℃~970℃**的操作温度
    - 防止启动时加入到槽内的熔融电解质凝固
    - 一般需要4~8天
  - 启动
    - 在焙烧终了后，融化**冰晶石**等溶剂形成铝电解所需的**熔融电解质**
    - 同时进一步加热炉内衬及清理炭渣，使电解槽的主要技术参数进入到电解所需的范围之内
  - 正常生产
    - 在930℃-970℃高温环境的**电解槽**内，1.92-1.94吨**氧化铝**作为原料与**冰晶石**、**氟化铝**等**氟盐溶剂**形成熔融状态的**电解质**，通入13500kw·h直流电进入电解槽，在阴极和阳极上发生电化学反应，得到电解产物
    - 阴极上是液体铝
    - 阳极上是
      - CO2（约75%~80%）
      - CO（20%~25%）
    - 氧化铝熔融于冰晶石形成的**电解质**的密度约为 2.1g/cm³ ，**铝液**密度为 2.3 g/cm³ ，两者因密度差而上下分层
    - **铝液**沉淀于电解槽底部，用真空抬包抽出后，经过净化和过滤，获得1吨99.5%~99.85%纯度的**原铝**

- 电解铝冶炼工艺属于**重型固定资产投资**
- 日常生产过程中，除了检修以及其他不可抗拒因素外，<font color="red">**连续性生产过程不会停止**</font>，每次重启电解槽的成本很高且会影响电解槽寿命

### 汽车铝消费

- 汽车轻量化发展的三大路径
  - 材料轻量化（替换轻质材料）
    - 目前汽车制造中用到的轻量化材料有：高强度钢、铝合金、镁合金、碳纤维复合材料、工程塑料等
    - 其中高强度钢和铝合金**性价比高**，特点明显，在汽车轻量化的材料中占比较高
    - 汽车制造使用的铝合金中，70%以上为**压铸铝合金**
      - 国内在**铝合金压铸工艺**上对比美国、日本还有一定差距，
      - **挤压铸造产品**存在质量稳定性差、废品率高等诸多问题
  - 结构轻量化（优化结构布局）
    - 对车身整体及零部件进行结构分析和优化
    - 改进车身结构，使部件**薄壁化**、**中空化**、**小型化**和**复合化**
    - 结构优化分为四种
      - 尺寸优化
        - 以零部件的尺寸作为优化的参数变量
        - 通过板材厚度、杆件横截面面积等尺寸关系，寻找最优设计参数的组合
      - 形状优化
        - 适当改变零部件的外形使结构更加均匀地受力
      - 形貌优化
        - 根据汽车各部件的强度要求和使用频率的不同
          - 在强度要求高的部位储存或增强材料
          - 在强度要求低的部位减薄或去除材料
      - 拓扑优化
        - 在给定设计空间的前提下，通过拓扑算法，对材料分布进行分析
        - 最终得到**最优化的动力传递路径**以达到尽可能多节省材料的目标
  - 工艺轻量化（改进生产工艺）
    - 是一座连接轻量化材料和轻量化结构的桥梁
    - 对于材料轻量化中运用最为广泛的高强度钢和铝合金，对应的制作工艺分别有
      - **热成型法**
        - 目前热成型工艺在汽车工业中应用最为广泛
      - **一体化压铸**
        - 压铸一体化优点在于一个铸件可以取代原本几十个零件，实现了零件高度集成化、模块化
        - 目前由于技术限制一体化压铸并未得到广泛应用
        - 在成本和轻量化驱动下，未来有望得到广泛应用，由此铝合金车身结构件中的份额将会随之提高

- 汽车的基本结构
  - 车身
    - 汽车车身主要组成
      - **车身结构件**
        - 主要目的是为车身起到支撑作用，类似于“人体骨骼”
        - 前后总成、A/B/C 柱、左右减震器悬挂部分、左右纵梁、横梁、上边梁等零部件
      - **车身覆盖件**
        - 起到美观和疏导气流的作用

      - **车身加强件**
  - 发动机
    - 发动机用铝主要为气缸体、气缸盖及活塞

  - 底盘

    - 底盘用铝量主要集中在四大系统的转向节、副车架、控制臂，制动钳等零部件上

## 未来供给增量：再生铝

- 再生铝不仅具有显著的经济优势，还能有效节约自然资源、保护生态环境
  - 氧化铝生产过程中会产生大量含金属废料的赤泥，目前暂无合适处理方法，大量赤泥的堆放严重影响生态环境，而再生铝的生产不涉及赤泥的产生，对环境保护有积极影响
    - 目前赤泥的综合利用率不足10%，大量的赤泥需要采用筑坝堆存的方式进行处置
- 据国际铝业协会和USGS数据显示
  - **欧洲**再生铝约占原铝和再生铝总量的**60%**
  - **美国**约为**83%**
  - **中国**再生铝占比仅为**17%**
  - **全球平均水平**为**34.3%**

# 铝价多空逻辑


## 利多因素


### 供给
- **能源供应**问题仍对产能释放有抑制
  - **电力**紧张（**限电**），导致电解铝生产成本增加，抑制产能，减少产量，会导致价格上涨
  - **干旱**导致水资源不足水电发电量下降，减少产能，铝价上涨
  - 动力煤价格上涨，导致铝加工成本上涨
- 预焙阳极价格上涨
  - 煤沥青、石油焦
  - 原材料**石油焦**价格跟随**原油**价格上涨而上涨，推升预焙阳极生产成本
- **库存不断去化**，会导致铝价上涨
  - **物流运输不畅**，会导致到货减少致使显性库存不断去化，铝价上涨
  - 发运周期延长，铝价上涨

### 需求
- 新能源板块消费维持高速增长；
  - **新能源车**（新能源车较传统燃油车用铝单耗高，助推汽车用铝量提升）
    - 车身板
    - 底盘
    - 发动机
  - **光伏**
    - 组件（背板）
    - 支架
- 国内稳地产、促内需政策发力带动传统消费修复；
  - **建筑铝型材**（开工率）
- 其它铝材开工率（上升）（不同铝材加工费不同，对于采用“铝材基价+加工费”模式结算的企业的利润有影响）
  - 工业铝型材
  - 铝板带
  - 铝线缆
  - 铝合金
- 全球加息节奏放缓，利率见顶；
  - 流动性宽松，会刺激经济，刺激对大宗商品需求，导致价格上涨
  - 流动性紧缩，将对包括铝在内的大宗商品价格不利
    - **加息**，对铝等大宗商品价格不利
    - **流动性收缩**可能导致**美元指数走强**，对铝等大宗商品价格不利

  

## 利空因素

### 供给

- 产能仍在释放，且仍有空间；
- 海外冶炼厂复产，叠加新产能投放；
- 水电站丰水期，导致产能释放，导致铝价格下跌


### 需求
- 海外经济放缓，出口回落；
  - 铝合金轮毂出口下降明显
    - 铝合金轮毂出口的下降与海外汽车产量下降有直接关系
- 国内疫情周期长，对经济负面影响持续
- **高通胀**引发经济衰退担忧加剧，导致铝价下跌



# 机会

- 产品的质量控制体系及**质量检测技术**差距较大
- **装备和工器具的设计制造**水平还较低
- 冶炼产能过剩，**高附加值的加工产品**短缺
  - 只有**西南铝**、**东北轻**和**南山**等少数厂家能够生产高合金化、高档次、高附加值的产品，总量不足50 万吨
  - 这类产品的国内消费量占板带总量的18%，即每年需求120万吨左右，大部分要依靠进口补充解决
  - 未来几年将从目前自给化率只有30%~50%提高到70%以上
- **铝液直接铸造铸轧加工技术**将得到快速发展
  - 用液态原铝铸坯进而直接生产板带材，可节电700 千瓦时/吨，减少烧损2.5%，综合成本降低约800 元/吨
  - 我国目前的比例只有25%，而发达国家已经接近70%
- **废铝**的加工分选技术有待于推进发展
  - 目前发达国家再生和原生铝的比例已接近5:5，而我国只有2:8
- **产业向中西部转移**步伐显著加快，集中度进一步提高

- 未来铝产业产能过剩的风险在加大，但电解铝厂商逐渐向下游深加工拓展的步伐也较快
